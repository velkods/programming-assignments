Easy:

1. How many unique customers are in the dataset?

```python
import pandas as pd
import datetime

inv = pd.read_csv('fct_invoice.csv')
cst = pd.read_json('dim_customer.json')

cst.count()

#There are 5,191 unique customers in the dataset.
```

2. What are the different categories of products available? How many unique categories are there?

```python
inv.category.unique()

#Eight categories: Clothing, Shoes, Books, Cosmetics, Food & Beverage, Toys, Technology, and Souvenirs.
```

3. Which payment method is the most popular? How many times was it used?

```python
inv['payment_method'].value_counts()

#Cash is the most popular payment method, with 44,447 uses.
```

Medium:

1. What are the three most popular categories, by total sales?

```python
inv.groupby('category')['quantity'].sum().sort_values()

#Clothing is the most popular category, with 103,558 units sold. 
#Cosmetics is the second most popular category, with 45,465 units sold. 
#Finally, Food & Beverage is the third most popular category, with 44,277 units sold. 
```

2. What are the total sales attributed to customers over the age of 45?

```python
ages = pd.merge(inv , cst, how = 'outer', left_on = 'customer_id', right_on ='id')

ages_filter = ages.loc[ages['age'] > 45]
ages_filter.groupby('age')['quantity'].sum().sum()

#97,629 (units of) sales are attributed to customers over the age of 45.
```

3. How is the data distributed across different invoice dates? Are there any seasonal trends or patterns? (Use a graph for this.)

```python
import matplotlib.pyplot as plt
import numpy as np

ages['invoice_date'] = pd.to_datetime(ages['invoice_date'], dayfirst=True)
ages['month-year'] = ages['invoice_date'].dt.to_period('m')
age_dates = ages.groupby('month-year')['quantity'].sum()
age_dates

age_dates.plot()
plt.xlabel('Date')
plt.ylabel('Sales (Total Quantity)')




#Sales have a tendency to dip in the first two months of each year (Jan/Feb of 2021-2022, Jan of 2023)
```

Hard:

1. Create a pivot table showing the breakdown of sales across these dimensions, in this order: category, decade age range (e.g., 10-19, 20-29, and so forth).

```python
age_range = []
for age_data in cst['age']:
        if age_data >= 10 and age_data < 20:
            age_range.append('10-19')
        elif age_data >= 20 and age_data < 30:
            age_range.append('20-29')
        elif age_data >= 30 and age_data < 40:
            age_range.append('30-39')
        elif age_data >= 40 and age_data < 50:
            age_range.append('40-49')
        elif age_data >= 50 and age_data < 60:
            age_range.append('50-59')
        elif age_data >= 60 and age_data < 70:
            age_range.append('60-69')
        elif age_data >= 70 and age_data < 80:
            age_range.append('70-79')
        elif age_data >= 80 and age_data < 90:
            age_range.append('80-89')
        elif age_data >= 90 and age_data < 100:
            age_range.append('90-99')
        else:
            age_range.append('No age data')

cst['age range'] = age_range

age_range_sales = inv.merge(cst, how = 'inner', left_on = 'customer_id', right_on = 'id')
pd.pivot_table(age_range_sales, values = 'quantity', index = 'category', columns = ['age range'], aggfunc = np.sum)
        
```
