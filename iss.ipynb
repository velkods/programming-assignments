```python
import matplotlib.pyplot as plt
import numpy as np
import requests
import time

longitude_data = []
latitude_data = []

i = 0
for i in range (0, 105):
    r = requests.get('http://api.open-notify.org/iss-now.json')
    longitude_data.append(r.json()['iss_position']['longitude'])
    latitude_data.append(r.json()['iss_position']['latitude'])
    i = i + 1
    time.sleep(10)
    if i == 100:
        print(longitude_data)
        print(latitude_data)
    
    
    
```

```python
longitude_data_float = [float(i) for i in longitude_data]
latitude_data_float = [float(i) for i in latitude_data]

x = np.array(longitude_data_float)
y = np.array(latitude_data_float)

plt.title("ISS Coordinates")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
plt.scatter(x, y)

plt.xticks(np.arange(min(longitude_data_float), max(longitude_data_float), 50))
plt.yticks(np.arange(min(latitude_data_float), max(latitude_data_float), 5))

plt.show()
```
